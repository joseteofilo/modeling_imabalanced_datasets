```python
# Python libraries are imported
import numpy as np
import pandas as pd
import pandas_profiling
import math
from rdkit.Chem import PandasTools
from IPython.display import HTML
```

```python
# load molecules
data1 = "../data/A2780.sdf"
data2 = "../data/CCRF-CEM.sdf"
data3 = "../data/DU-145.sdf"
data4 = "../data/HCT-15.sdf"
data5 = "../data/KB.sdf"
data6 = "../data/LoVo.sdf"
data7 = "../data/PC-3.sdf"
data8 = "../data/SK-OV-3.sdf"
```

```python
# Convert the SDF file to a dataframe with smiles 
compound_df1 = PandasTools.LoadSDF(data1, smilesName='smiles', includeFingerprints=False)
compound_df2 = PandasTools.LoadSDF(data2, smilesName='smiles', includeFingerprints=False)
compound_df3 = PandasTools.LoadSDF(data3, smilesName='smiles', includeFingerprints=False)
compound_df4 = PandasTools.LoadSDF(data4, smilesName='smiles', includeFingerprints=False)
compound_df5 = PandasTools.LoadSDF(data5, smilesName='smiles', includeFingerprints=False)
compound_df6 = PandasTools.LoadSDF(data6, smilesName='smiles', includeFingerprints=False)
compound_df7 = PandasTools.LoadSDF(data7, smilesName='smiles', includeFingerprints=False)
compound_df8 = PandasTools.LoadSDF(data8, smilesName='smiles', includeFingerprints=False)
```

```python
# Check the number of entries
print(len(compound_df1))
print(len(compound_df2))
print(len(compound_df3))
print(len(compound_df4))
print(len(compound_df5))
print(len(compound_df6))
print(len(compound_df7))
print(len(compound_df8))
```

```python
# Check clolumns
list(compound_df1)
list(compound_df2)
list(compound_df3)
list(compound_df4)
list(compound_df5)
list(compound_df6)
list(compound_df7)
list(compound_df8)
```

```python
# Float column 'pIC50'
compound_df1["pIC50"] = compound_df1["pIC50"].astype(float)
compound_df2["pIC50"] = compound_df2["pIC50"].astype(float)
compound_df3["pIC50"] = compound_df3["pIC50"].astype(float)
compound_df4["pIC50"] = compound_df4["pIC50"].astype(float)
compound_df5["pIC50"] = compound_df5["pIC50"].astype(float)
compound_df6["pIC50"] = compound_df6["pIC50"].astype(float)
compound_df7["pIC50"] = compound_df7["pIC50"].astype(float)
compound_df8["pIC50"] = compound_df8["pIC50"].astype(float)
```

```python
# Round values in the column 'pIC50'
decimals = 2 
compound_df1["pIC50"] = [round(x, decimals) for x in compound_df1["pIC50"]]
compound_df2["pIC50"] = [round(x, decimals) for x in compound_df2["pIC50"]]
compound_df3["pIC50"] = [round(x, decimals) for x in compound_df3["pIC50"]]
compound_df4["pIC50"] = [round(x, decimals) for x in compound_df4["pIC50"]]
compound_df5["pIC50"] = [round(x, decimals) for x in compound_df5["pIC50"]]
compound_df6["pIC50"] = [round(x, decimals) for x in compound_df6["pIC50"]]
compound_df7["pIC50"] = [round(x, decimals) for x in compound_df7["pIC50"]]
compound_df8["pIC50"] = [round(x, decimals) for x in compound_df8["pIC50"]]
```

```python
# Remove columns
compound_df1 = compound_df1.drop(['ROMol'], axis=1)
compound_df2 = compound_df2.drop(['ROMol'], axis=1)
compound_df3 = compound_df3.drop(['ROMol'], axis=1)
compound_df4 = compound_df4.drop(['ROMol'], axis=1)
compound_df5 = compound_df5.drop(['ROMol'], axis=1)
compound_df6 = compound_df6.drop(['ROMol'], axis=1)
compound_df7 = compound_df7.drop(['ROMol'], axis=1)
compound_df8 = compound_df8.drop(['ROMol'], axis=1)
```

## Exploring data set cell line "A2780" - Ovarian carcinoma cells

```python
# Exploring data
from pandas_profiling import ProfileReport
profile = ProfileReport(compound_df1, title='Pandas Profiling Report', html={'style':{'full_width':True}})
profile
```

```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

pIC50s = np.array(compound_df1["pIC50"])
# Remove some dirty data from the dataset
pIC50s = [pIC50 for pIC50 in pIC50s if pIC50 != '']
n, bins, patches = plt.hist(pIC50s, 100, facecolor='green', alpha=0.75)
plt.xlabel('Measured pIC50')
plt.ylabel('Number of compounds')
plt.title(r'Histogram of pIC50 Values')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set_style('white')
```

```python
sns.distplot(compound_df1["pIC50"])
plt.xlabel("pIC50")
plt.ylabel('probability')
```
